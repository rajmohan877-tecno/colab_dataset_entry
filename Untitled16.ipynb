```python
import edvart
import pandas as pd
from edvart import Verbosity

import plotly.offline as py
py.init_notebook_mode()
```

```python
def dataset_AI_index_db():
    return pd.read_csv('AI_index_db.csv')
```

```python
dataset = dataset_AI_index_db()

dataset_name = "AI_index_db"
dataset_description = "Dataset that contains data for AI_index_db."
```

```python
dataset.head()
```

```python
report = edvart.DefaultReport(
    dataset,
    verbosity=Verbosity.LOW,
    columns_univariate_analysis=['Income group', 'Region', 'Cluster'],
    columns_bivariate_analysis=['Infrastructure', 'Operating Environment', 'Research', 'Government Strategy'],
    groupby='Total score',
)
```

```python
report.show()
```

```python
report.export_notebook(
    notebook_filepath='exported-notebook.ipynb',
    dataset_name=dataset_name,
    dataset_description=dataset_description,
)
```

```python
report.export_html(
    html_filepath='exported-report.html',
    dataset_name=dataset_name,
    dataset_description=dataset_description,
)
```
