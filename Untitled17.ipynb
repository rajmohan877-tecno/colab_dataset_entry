```python
%pip install edvart
import edvart
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import plotly.offline as py
py.init_notebook_mode()
```

```python
dataset_ts = edvart.example_datasets.dataset_global_temp()
```

```python
report_ts = edvart.DefaultTimeseriesReport(
    dataset_ts,
    # Monthly data -> analyze yearly seasonality
    sampling_rate=12,
    stft_window_size=24,
)
```

```python
report_ts.show()
```

```python
report_ts.export_notebook(
    "exported-notebook-time-series.ipynb",
    dataset_name="Global Temperature",
    dataset_description="Monthly global temperature anomaly form 1880 until 2016"
)
```

```python
report_ts.export_html(
    "exported-report-time-series.html",
    dataset_name="Global Temperature",
    dataset_description="Monthly global temperature anomaly form 1880 until 2016"
)
```

## Basic report

```python
dataset = edvart.example_datasets.dataset_titanic()

dataset_name = "Titanic"
dataset_description = "Dataset that contains data for 891 of the real Titanic passengers."
```

```python
dataset.head()
```

```python
report = edvart.DefaultReport(
    dataset,
    verbosity=Verbosity.LOW,
    columns_univariate_analysis=['Name', 'Age', 'Pclass'],
    columns_bivariate_analysis=['Survived', 'Sex', 'Fare', 'Pclass'],
    groupby='Survived',
)
```

```python
report.show()
```

```python
report.export_notebook(
    notebook_filepath='exported-notebook.ipynb',
    dataset_name=dataset_name,
    dataset_description=dataset_description,
)
```

```python
report.export_html(
    html_filepath='exported-report.html',
    dataset_name=dataset_name,
    dataset_description=dataset_description,
)
```

```python
import edvart
from edvart import Verbosity

import plotly.offline as py
py.init_notebook_mode()
```

```python
def dataset_AI_index_db():
    return pd.read_csv('AI_index_db.csv')
```

```python
dataset = dataset_AI_index_db()

dataset_name = "AI_index_db"
dataset_description = "Dataset that contains data for AI_index_db."
```

```python
dataset.head()
```

```python
report = edvart.DefaultReport(
    dataset,
    verbosity=Verbosity.LOW,
    columns_univariate_analysis=['Income group', 'Region', 'Cluster'],
    columns_bivariate_analysis=['Infrastructure', 'Research', 'Development', 'Government Strategy'],
    groupby='Country',
)
```

```python
report.show()
```

```python
report.export_notebook(
    notebook_filepath='exported-notebook.ipynb',
    dataset_name=dataset_name,
    dataset_description=dataset_description,
)
```

```python
report.export_html(
    html_filepath='exported-report.html',
    dataset_name=dataset_name,
    dataset_description=dataset_description,
)
```
