```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Numerical Data plot

## Distribution plot

```python
sns.set(style = 'darkgrid', rc = {'figure.figsize':(10, 6)}, font_scale = 1.5)
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot
```

```python
sns.distplot(tips['total_bill'], bins = 20)
```

```python
sns.distplot(tips['total_bill'], kde = False)
```

```python
sns.distplot(tips['total_bill'], hist = False, vertical= True, color = 'red')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde', color = 'red')
```

## Pair Plot

```python
sns.pairplot(tips)
```

```python
tips.head()
```

```python
sns.pairplot(tips, hue = 'sex')
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

# Categorical Ploting

```python
tips.head()
```

## Barplot

```python
sns.barplot(x ='sex' , y = 'total_bill' , data = tips)
```

```python
sns.barplot(x ='sex' , y = 'total_bill' , data = tips, hue = 'smoker')
```

```python
sns.barplot(x ='sex' , y = 'total_bill' , data = tips, hue = 'day')
```

```python
sns.barplot(x ='sex' , y = 'total_bill' , data = tips, hue = 'day', estimator = np.std)
```

## Count Plot

```python
sns.countplot(x = 'sex', data = tips)
```

```python
sns.countplot(x = 'sex', data = tips, hue = 'time')
```

```python
sns.countplot(x = 'sex', data = tips, hue = 'time', color = 'green')
```
